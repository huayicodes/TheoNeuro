```python
import numpy as np
import matplotlib.pyplot as plt

# set time
T = 10
dt =  0.01
t = np.arange(0, T, dt)

## Unconmment the two lines below to check x, y relationship for 0<k<2. 
kstep = 0.01
k = np.arange(0+kstep, 2, kstep)

# Assume x & y starts at 0.001, through running the simulation, I have found that the largest value of k for these equations 
# to have oscillations is k = 0.83. 
#kstep = 0.01
#k = np.arange(0.80, 0.85, kstep)

plt.figure()
for j in k: 
    x = np.zeros(len(t))
    y = np.zeros(len(t))
    # assume the initial values of x & y are small. 
    x[0] = 0.001
    y[0] = 0.001
    for i in np.arange(1,len(t)):
        dx = dt * (- x[i-1] * (x[i-1] - 1) * (x[i-1] + 1) + y[i-1])
        dy = dt * j * (2*x[i-1] + y[i-1])
        # update x & y
        x[i] = x[i-1] + dx
        y[i] = y[i-1] + dy
    plt.plot(x,y)
    plt.ylabel('y')
    plt.xlabel('x')
    plt.title(('k= %f'%j ) )
    plt.show()
```
