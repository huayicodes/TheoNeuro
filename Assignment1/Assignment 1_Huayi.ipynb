Qn3. Build an integrate-and-fire model neuron, 

$τ_m\frac{dv}{dt} = V_{rest} − V + R_mI_e$

With $V_{rest} = V_{reset} = −65 \ mV, \ V_{th} = −50 \ mV, τ_m = 10 \ ms$, and $R_m = 10 \ MΩ$. Reset the potential to $V =V_{reset}$ whenever it goes to or above $V_{th}$, and then the neuron fires an action potential.
Apply different levels of current $I_e$, count spikes over a suitable period of time to compute firing rates, and plot these rate as a function of $R_mI_e$. Compare your results to the analytic formula

$r =(τ_m \ln \frac{R_mI_e + V_{rest} − V_{reset}}{R_mI_e + V_{rest} − V_{th}})^{−1}$
.

```python
import numpy as np
import matplotlib.pyplot as plt



def InF(Ie):
    # Set up the parameters 
    V_rest = V_reset = -65.0
    V_th = -50.0
    tau_m= 10.0
    R_m = 10.0
    
    # set time
    dt =  0.001
    T = 10.0 #duration in s
    t = np.arange(0, T, dt)
    
    # initiate varibles
    V = np.zeros(len(t))
    #initial number of spike is 0
    f_count = np.zeros(1) 
    #start from resting potential 
    V[0] = V_rest
    for i in np.arange(1, len(t)):
        # by Euler approximation, solution for "tau_m * (dv/dt) = V_rest - V + R_m * Ie" is: 
        dv = (V_rest - V[i-1] + R_m * Ie)/tau_m * dt
        # update V
        V[i] = V[i-1] + dv            
        # reset when fires 
        if V[i] > V_th:
            V[i] = V_reset
            f_count += 1 
    f_rate = f_count/ T #Hz
    return (R_m*Ie,f_rate)

Ie = np.arange(0,100,5)
f_rates = np.zeros(len(Ie))
R_mIe = np.zeros(len(Ie))
for i,j in enumerate(Ie): 
    R_mIe[i],f_rates[i] = InF(j)

simulate = plt.scatter(R_mIe, f_rates, label='simulate')
plt.xlabel('Rm*Ie (mV)')
plt.ylabel('firing rates (Hz)')


V_rest = V_reset = -65.0
V_th = -50.0
tau_m= 10.0
R_m = 10.0

# compare results to the analytical formula 
with np.errstate(divide='ignore'):
    r = 1/(tau_m * np.log((R_mIe+V_rest- V_reset)/(R_mIe+V_rest- V_th)))
analytical = plt.scatter(R_mIe, r, color = 'r',  label='analytical')
plt.legend(handles=[simulate, analytical], loc=2)
plt.show()

# the stimulation and the analytical resutls are very similar. 
```

Qn4. Build a Hodgkin-Huxley model neuron by numerically integrating the equations for $V$, $m$, $h$,and $n$:

$c_m\frac{dv}{dt} = −i_m + I_e/A$

where: $i_m = \bar{g}_L(V − E_L) + \bar{g}_Kn^4(V − E_K) + \bar{g}_Nam^3h(V − E_Na)$

and: $τ_n(V)\frac{dv}{dt}= n_∞(V) − n \\ τ_n(V) =\frac{1}{α_n(V) + β_n(V)}$

and: $n_∞(V) =\frac{α_n(V)}{α_n(V) + β_n(V)}$

and similar equations for m and h, with
$α_n =\frac{.01(V + 55)}{1 − exp(−.1(V + 55))} \ \ \ β_n = 0.125 exp(−0.0125(V + 65))\\
α_m =\frac{.1(V + 40)}{1 − exp(−.1(V + 40))} \ \ \ β_m = 4 exp(−.0556(V + 65))\\
α_h = .07 exp(−.05(V + 65)) \ \ \ β_h = 1/(1 + exp(−.1(V + 35)))$

In these equations, time is in ms and voltage is in mV. Take $c_m = 10 \ nF/mm^2$ and, as initial values, take: V = −65 mV, m = 0.0529, h = 0.5961, and n = 0.3177. The maximal conductances and reversal potentials used in the model are $\bar{g}_L = 0.003 \  mS/mm^2 $, $\bar{g}_K = 0.36 \ mS/mm^2$, $\bar{g}Na = 1.2 \ mS/mm^2$, $E_L = -54.387 \ mV$, $E_K = -77 \ mV$ and $E_{Na} = 50 \ mV$. Make your integration step size small enough to get accurate results.

a) Use an external current with $I_e/A = 200 \ nA/mm^2$ and plot V, m, h, and n as functions of time for a suitable interval. 

```python
def HH(Ie_ov_A = None):
    # setting default value of Ie_ov_A. 
    if Ie_ov_A == None:
        Ie_ov_A = 200
    else:
        Ie_ov_A = Ie_ov_A 
    
    # variables related to n
    an=lambda v: 0.01*(v + 55)/ (1 - np.exp(-0.1*(v+55))) if v != -55 else 0.01
    bn= lambda v: 0.125 * np.exp(- 0.0125*(v + 65))
    n_inf = lambda v: an(v)/(an(v) + bn(v))
    tau_n = lambda v: 1/(an(v) + bn(v))
    
    # variables related to m
    am=lambda v: 0.1*(v + 40)/ (1 - np.exp(-0.1*(v + 40))) if v != -40 else 0.1
    bm= lambda v: 4* np.exp(- 0.0556 * (v + 65))
    m_inf = lambda v: am(v)/(am(v) + bm(v))
    tau_m = lambda v: 1/(am(v) + bm(v))

    # variables related to h
    ah=lambda v: 0.07 * np.exp(- 0.05 * (v + 65)) 
    bh= lambda v: 1/(1 + np.exp(- 0.1*(v + 35)))
    h_inf = lambda v: ah(v)/(ah(v) + bh(v))
    tau_h = lambda v: 1/(ah(v) + bh(v))


    T = 100 #ms
    cm = 10 
    EL = -54.387
    EK = -77 
    ENa = 50
    # to get the unit right
    gL_bar = 0.003 * 1000 # in nS/mm^2
    gK_bar = 0.36 * 1000 #  in nS/mm^2
    gNa_bar = 1.2 * 1000 #  in nS/mm^2

    # initial values 
    vi = -65
    mi = 0.0529
    hi = 0.5961
    ni = 0.3177

    # set time
    dt =  0.01
    t = np.arange(0, T, dt)

    v = np.zeros(len(t))
    m = np.zeros(len(t))
    h = np.zeros(len(t))
    n = np.zeros(len(t))

    v[0] = vi
    m[0] = mi
    h[0] = hi
    n[0] = ni
    cnt = 0

    for i in np.arange(1,len(t)):
        n[i] =  n_inf(v[i-1]) + (n[i-1] - n_inf(v[i-1])) * np.exp(-dt/tau_n(v[i-1]))
        m[i] =  m_inf(v[i-1]) + (m[i-1] - m_inf(v[i-1])) * np.exp(-dt/tau_m(v[i-1]))
        h[i] =  h_inf(v[i-1]) + (h[i-1] - h_inf(v[i-1])) * np.exp(-dt/tau_h(v[i-1]))
        #simplify the equation cm * dv/dt = - Im + Ie/A to the form cm * dv/dt = coA (-v) + coB. 
        # (cm/ coA) * dv/dt = -v  + coB/coA. Then, v can be integrated by the exponential method. 
        coA = (gL_bar + gK_bar* np.power(n[i-1],4) + gNa_bar*np.power(m[i-1],3)*h[i-1])
        coB = (gL_bar* EL + gK_bar* np.power(n[i-1],4)* EK + gNa_bar*np.power(m[i-1],3)*h[i-1]*ENa) + Ie_ov_A 
        v[i] = coB/coA + (v[i-1] - coB/coA) * np.exp(-dt/(cm/coA))
        # detect spike when v[i] turn positive for the first time
        if (v[i-1]<0 and v[i]>0):
            cnt += 1
    if cnt == 0: 
        frate = 0
    else: 
        frate = cnt/(T/1000) # from "ms" to "s" to get Hz
    return (m,h,n,v,t,frate)

m,h,n,v,t,frate = HH()
plt.figure()
plt.plot(t,v)
plt.xlabel("t (s)")
plt.ylabel(" V (mV)")
plt.show()

plt.figure()
plotm, = plt.plot(t,m, color = 'm',  label='m' )
ploth, = plt.plot(t,h, color = 'r',  label='h' )
plotn, = plt.plot(t,n, color = 'g',  label='n' )
plt.xlabel('t (s)')
plt.legend(handles=[plotm, ploth, plotn], loc=2)
plt.show()
```

Qn5: Use the Hodgkin-Huxley model you built for problem 4 to construct a multiple compartment model of an axon. Built the axon out of 100 compartments, each of which has a radius of 2 µm and a length of 100 µm. Each compartment should contain the Hodgkin-Huxley conductances used in problem 4. Inject a current pulse sufficient to evoke a single action potential into the first compartment of the cable. All the other compartments should receive no external current. Use the Euler integration method or one of the build-in Matlab integration routines with a time step no bigger than around 0.01 ms.

a) What is the propagation speed of the resulting action potential.

```python
def axon(r, l): 
    Ie_ov_A = 80 # current that initiate a single action potential

    # variables related to n
    an=lambda v: 0.01*(v + 55)/ (1 - np.exp(-0.1*(v+55))) 
    bn= lambda v: 0.125 * np.exp(- 0.0125*(v + 65))
    n_inf = lambda v: an(v)/(an(v) + bn(v))
    tau_n = lambda v: 1/(an(v) + bn(v))

    # variables related to m
    am=lambda v: 0.1*(v + 40)/ (1 - np.exp(-0.1*(v + 40))) 
    bm= lambda v: 4* np.exp(- 0.0556 * (v + 65))
    m_inf = lambda v: am(v)/(am(v) + bm(v))
    tau_m = lambda v: 1/(am(v) + bm(v))

    # variables related to h
    ah=lambda v: 0.07 * np.exp(- 0.05 * (v + 65)) 
    bh= lambda v: 1/(1 + np.exp(- 0.1*(v + 35)))
    h_inf = lambda v: ah(v)/(ah(v) + bh(v))
    tau_h = lambda v: 1/(ah(v) + bh(v))


    T = 200
    cm = 10
    EL = -54.387
    EK = -77 
    ENa = 50
    # to get the unit right
    gL_bar = 0.003 * 1000 # in nS/mm^2
    gK_bar = 0.36 * 1000 #  in nS/mm^2
    gNa_bar = 1.2 * 1000 #  in nS/mm^2

    # initial values 
    vi = -65
    mi = 0.0529
    hi = 0.5961
    ni = 0.3177

    # set time
    dt =  0.01
    t = np.arange(0, T, dt)

    # parameters for the axon
    N = 100
    
    gA_bar = (1/(2*r*(l**2))) * (10**6) # nS/mm^2

    v_axon = np.zeros([len(t), N])
    m = np.zeros([len(t), N])
    h = np.zeros([len(t), N])
    n = np.zeros([len(t), N])
    istart = np.zeros(len(t))
    iend = np.zeros(len(t))
    v_axon[0, :] = vi
    m[0, :] = mi
    h[0, :] = hi
    n[0, :] = ni
    
    for i in np.arange(1,len(t)):
        n[i,:]  =  n_inf(v_axon[i-1,:]) + (n[i-1,:] - n_inf(v_axon[i-1,:])) * np.exp(-dt/tau_n(v_axon[i-1,:]))
        m[i,:]  =  m_inf(v_axon[i-1,:]) + (m[i-1,:] - m_inf(v_axon[i-1,:])) * np.exp(-dt/tau_m(v_axon[i-1,:]))
        h[i,:]  =  h_inf(v_axon[i-1,:]) + (h[i-1,:] - h_inf(v_axon[i-1,:])) * np.exp(-dt/tau_h(v_axon[i-1,:]))
        Im = gL_bar*(v_axon[i-1,:]-EL) + gK_bar* np.power(n[i-1,:],4)*(v_axon[i-1,:]-EK) + gNa_bar*np.power(m[i-1,:],3)*h[i-1,:]*(v_axon[i-1,:]-ENa)

        ## first compartment  
        dv1 = dt/cm*(-Im[0] + Ie_ov_A + gA_bar*(v_axon[i-1,1] -v_axon[i-1,0]))
        v_axon[i,0] = v_axon[i-1,0] + dv1
        # time for action potential to cross 0 at first compartment 
        if (v_axon[i-1,0]<0 and v_axon[i,0]>0):
            istart[i-1] = i*dt
 
        # #the other compartments 
        dv = dt/cm*(-Im[1:-1] + gA_bar*(v_axon[i-1,:-2]-v_axon[i-1,1:-1]) + gA_bar* (v_axon[i-1, 2:] -v_axon[i-1,1:-1]))
        v_axon[i,1:-1] = v_axon[i-1,1:-1] + dv 
        
        ## last compartemnt 
        dvN = dt/cm*(-Im[-1] + gA_bar*(v_axon[i-1,-2] -v_axon[i-1,-1]))
        v_axon[i,-1] = v_axon[i-1,-1] + dvN
        
        # time for action potential to cross 0 at last compartment 
        if (v_axon[i-1,-1]<0 and v_axon[i,-1]>0):
            iend[i-1] = i*dt
          
       
    cspeed = l*N/(iend[np.nonzero(iend)][0] - istart[np.nonzero(istart)][0])/1000 #mm/ms
    return(cspeed)

r = 2
l = 100
cspeed= axon(r,l)
print("when r = 2 um, conduction speed is %f mm/ms" %cspeed)
```

$τ_m\frac{dv}{dt} = V_{rest} - V + \frac{(V-V_{rest})^2}{∆_V} − U + R_mI_e,$

and

$τ_U\frac{dU}{dt} = b(V - V_{rest}) - U,$

with the additional condition that when $V ≥ V_{max}$ it gets reset to $V =V_{reset}$ and U gets
incremented by U → U +d. To start, use (in units of ms, mV, MΩ and nA) $τ_m =1.67 $,
$τ_U = 50$, $∆_V = 15$, $V_{rest} = −70$, $V_{max} = 0$, $V_{reset} = −50$, b = 0.33, d = 3, $R_m$ = 10 and
$I_e$=0.63.

a) Make plots of V versus t and U versus t for a long enough time period to show
what is happening.

The neuron is bursting periodically! 

```python
def Itz(tau_U=None, b =  None): 
    # set parameters 
    tau_m = 1.67
    delta_V = 15 
    V_rest = -70
    V_max = 0
    
    V_reset = -50 
    
    # set default values 
    if tau_U  == None: 
        tau_U = 50 
    else:
        tau_U = tau_U 
    if b == None: 
        b = 0.33   
    else:
        b = b 
    d = 3 
    
    R_m = 10 
    I_e = 0.63    

    T = 800

    # set time
    dt =  0.025
    t = np.arange(0, T, dt)

    # intiate V & U
    V = np.zeros(len(t))
    U = np.zeros(len(t))
    # starting from rest
    V[0] = V_rest

    for i in np.arange(1,len(t)):
        dv = dt/tau_m * (V_rest - V[i-1] + np.power((V[i-1] - V_rest ),2) / delta_V - U[i-1] + R_m * I_e)
        dU = dt/tau_U * (b * (V[i-1] - V_rest ) - U[i-1])
        # update V & U
        V[i] = V[i-1] + dv
        U[i] = U[i-1] + dU
        # reset when >= Vmax 
        if V[i] >= V_max:
            V[i] = V_reset
            U[i] = U[i] + d
    return (U,V,t)

U,V,t = Itz()

plt.figure()
plotV, = plt.plot(t,V, label='V' )
plotU, = plt.plot(t,U, label='U' )
plt.xlabel('t (s)')
plt.ylabel('V(mV)')
plt.legend(handles=[plotV, plotU], loc=1)
plt.show()
```

c) Explain the trajectory you see in b in terms of the nullclines and other features of
the model. For example, why does the trajectory cling to one of the nullclines over
part of the trajectory.

The interceptions of the two nullclines are the two fix points, representing cases that the tendency for both V & U to change is 0. But the nature of the two fixed points are different. The one on the left is a stable fix point. If we put a point in the phase plane left to this fix point, it will go towards the lowest point on dV/dt = 0, and upwards towards higher positions on dU/dt = 0, and eventually settles around the fixed point. Thus the trajectory on the left cling onto the nullcline dV/dt =0. The fix point on the right is unstable, and both nuclines have a rising trend, and a point near that area will not stay around this fix point. In this case, we set the bounding condition to $V_{max} = 0$, and thus the value of V will be bounced back and forth, as shown in the plot and resulted in repetitive bursts of firing.

Qn7. Simulate the pair of equations

$\frac{dx}{dt} = −x(x − 1)(x + 1) + y \ \ \ \ \ \ \frac{dy}{dt} = −k(2x + y)$

with different values of k in the range $ 0 < k < 2 $.

a) Find the largest value of k for which these equations produce oscillations.

```python
# set time
T = 10
dt =  0.01
t = np.arange(0, T, dt)

##  uncomment the two lines below to check x, y relationship for 0<k<2. 
#kstep = 0.01
#k = np.arange(0+kstep, 2, kstep)

# Answer: the maximum value for k to have oscillations is k<2, for 0<k<2. 
k = [1.98,1.99]

plt.figure()
for j in k: 
    x = np.zeros(len(t))
    y = np.zeros(len(t))
    # assume the initial values of x & y are small. 
    x[0] = 0.001
    y[0] = 0.001
    for i in np.arange(1,len(t)):
        dx = dt * (- x[i-1] * (x[i-1] - 1) * (x[i-1] + 1) + y[i-1])
        dy = dt * -j * (2*x[i-1] + y[i-1])
        # update x & y
        x[i] = x[i-1] + dx
        y[i] = y[i-1] + dy
    plt.plot(x,y)
    plt.ylabel('y')
    plt.xlabel('x')
    plt.title(('k= %f'%j ) )
    plt.show()

```

b) Compute this value analytically by considering the fixed point x = y = 0 and
finding the value of k where this state makes the transition from being stable to being
unstable. When this state is stable, is the approach to the fixed point exponential
or oscillatory? When the state is unstable, is the motion away from the fixed point
exponential or oscillatory?

Answer: Assume we are considering values very close to the fixed point, the behavior of the system can be approximated by $f(x) = J|_{x=x^*} (x-x^*)$. Solving the Jacobian matrix, I get: $J = (1, \ 1; -2k, \ -k)$. Then, solve for the eigen values of J. $ λ_{1,2} = \frac {(1-k) \pm \sqrt{((k-3)^2 - 8))}} {2}$. When $((k-3)^2 - 8)) >= 0$, as $ 0 < k < 2$, thus $ 0 < k <= 3 - \sqrt8$. Then λ > 0, the system is unstable, the escape from fixed point is exponential as there's no imaginary part. If $ k > 3 - \sqrt8$, λ is complex, and if k < 1, then the real part of λ > 0, system will be unstable, and the escape will be oscillatory. But if k>1, the system will be stable, and the approach to fix point is oscillatory.

Qn8: Construct an integrate-and-fire model responding to a "noisy" input representing the
in vivo environment. This model is based on the equation

$τ_m\frac{dv}{dt} = V_{rest} - V(t) + η(t) + R_mI_e$

with $τ_m = 10 ms$ and $V_{rest} = −56 mV$. The threshold and reset potentials for the
model are $V_{th} = −54 mV$ and $V_{reset} = −80 mV$. η is white-noise given by

$η = σV \sqrt \frac{2τ_m}{∆t}randn(1, length(t))$

where ∆t is the time step size in your program, $σ_V$ is a parameter (see below), randn
is the matlab random number generator for a normal distribution, and t is the vector
of times in your program.

a) Set $I_e = 0$ and turn off the spike generation mechanism in your model. Plot the
standard deviation of the membrane potential fluctuations that arise from different
σV values in the range 0 ≤ $σ_V$ ≤ 10 mV.

```python
def HHn(sigma_V, RmIe, spiking = None):
    # set parameters
    if spiking == None: 
        spiking == False 
        
    tau_m = 10.0 #ms
    V_rest = -56
    V_th = -54
    V_reset = -80
    
    # set time
    T =20.0 #duration in seconds. 
    dt =  0.01
    t = np.arange(0, T, dt)

    # initiate V
    V = np.zeros(len(t))
    # starting from rest
    V[0] = V_rest
    
    # initiate f_rates and std_V
    f_rate = np.zeros(len(sigma_V))
    std_V = np.zeros(len(sigma_V))
    
    for j, q in enumerate(sigma_V): 
        eta = q* np.power(2*tau_m/dt,0.5) * np.random.rand(len(t))
        f_count = 0
        for i in np.arange(1,len(t)): 
            dv = dt/tau_m*(V_rest  - V[i-1] + eta[i-1] + RmIe)
            V[i] = V[i-1] + dv
            # reset when fires
            if spiking == True: 
                if V[i] > V_th:
                    V[i] = V_reset
                    f_count += 1 
        f_rate[j] = f_count/ T #Hz
        std_V[j] = np.std(V)
    return(std_V, f_rate)


sigmaVstep = 0.1
sigma_V = np.arange(0,10+sigmaVstep,sigmaVstep )
sstd_V, _ = HHn(sigma_V, 0, None)
plt.plot(sigma_V, std_V )
plt.ylabel('stdV (mV)')
plt.xlabel('sigmaV (mV)')

plt.show()
```

```python
# generate RmIe & sigma_V
RmIe = np.arange(-300,300,5)
sigma_V = np.asarray([2,6,10])

# initiate firing rates
frates = np.zeros([len(RmIe), len(sigma_V)])
for u, v in enumerate (RmIe): 
    _, frates[u,:] = HHn(sigma_V, v, True)
    
plt.plot(RmIe,frates)
plt.ylabel('fring rates (hZ)')
plt.xlabel('RmIe (mV)')
plt.legend(['sigmaV = 2 ','sigmaV = 6', 'sigmaV = 10'], loc=2)
plt.show()
```
